```python
#General
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Scikit Preprocessing
from sklearn.impute import SimpleImputer
from sklearn.base import BaseEstimator,TransformerMixin
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.utils.validation import check_is_fitted
from sklearn.compose import ColumnTransformer
# Scikit Models/Model Selection
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
#Validación/Metrics
from sklearn.model_selection import cross_val_score,cross_val_predict
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import roc_curve,roc_auc_score
## Hiperparámetros
from sklearn.model_selection import GridSearchCV
```

```python
train_set = pd.read_csv('/mnt/d/ML/kaggle/titanic/train.csv')
test_set = pd.read_csv('/mnt/d/ML/kaggle/titanic/test.csv')
```

```python
titanic = train_set.drop("Survived", axis=1)
titanic_labels = train_set["Survived"].copy()
titanic_labels = titanic_labels.to_numpy()
```

# Relleno de los valores que faltan / Creación de atributos nuevos

```python
class GroupImputer(BaseEstimator,TransformerMixin):
    '''
    Class used for imputing missing values in a pd.DataFrame using either mean or median of a group.
    
    Parameters
    ----------    
    group_cols : list
        List of columns used for calculating the aggregated value 
    target : str
        The name of the column to impute
    metric : str
        The metric to be used for remplacement, can be one of ['mean', 'median']
    
    Returns
    -------
    X : array-like
        The array with imputed values in the target column
    '''
    def __init__(self,group_cols,target,metric='median'):
        assert metric in ['mean','median'], 'Unrecognized value for metric, should be mean/median.'
        assert type(group_cols) == list,'group_cols should be a list of columns.'
        assert type(target) == str, 'target should be a string.'
        
        self.group_cols = group_cols
        self.target = target
        self.metric = metric
    
    def fit(self,X,y=None):
        assert pd.isnull(X[self.group_cols].any(axis=None)) == False, 'There are missing values in group_cols'
        
        impute_map = X.groupby(self.group_cols)[self.target].agg(self.metric).reset_index(drop=False)
        self.impute_map_ = impute_map
        
        return self
    def transform(self,X,y=None):
        check_is_fitted(self,'impute_map_')
        
        X = X.copy()
        
        for index,row in self.impute_map_.iterrows():
            ind = (X[self.group_cols] == row[self.group_cols]).all(axis=1)
            X.loc[ind,self.target] = X.loc[ind,self.target].fillna(row[self.target])
        return X.values
```

```python
gimp = GroupImputer(group_cols=['Pclass','Sex'],target='Age')
titanic = pd.DataFrame(gimp.fit_transform(titanic),columns=titanic.columns)

test_set = pd.DataFrame(gimp.fit_transform(test_set),columns=test_set.columns)
```

```python
titanic['Deck'] = titanic['Cabin'].apply(lambda cabin: cabin[0] if pd.notnull(cabin) else 'U')
test_set['Deck'] = test_set['Cabin'].apply(lambda cabin: cabin[0] if pd.notnull(cabin) else 'U')
```

```python
titanic['Deck'] = titanic['Deck'].replace(['A','B','C'],'ABC')
titanic['Deck'] = titanic['Deck'].replace(['D','E'],'DE')
titanic['Deck'] = titanic['Deck'].replace(['F','G'],'FG')
titanic['Deck'] = titanic['Deck'].replace(['U','T'],'U') # Recordar esto!

test_set['Deck'] = test_set['Deck'].replace(['A','B','C'],'ABC')
test_set['Deck'] = test_set['Deck'].replace(['D','E'],'DE')
test_set['Deck'] = test_set['Deck'].replace(['F','G'],'FG')
test_set['Deck'] = test_set['Deck'].replace(['U','T'],'U') # Recordar esto!
```

```python
titanic['Family'] = titanic['Parch'] + titanic['SibSp']
test_set['Family'] = test_set['Parch'] + test_set['SibSp']

# Convertimos a Family en una variable categórica, de acuerdo a si es o no >=3. También distinguimos si iban solas con u '0'
titanic['Family'] = titanic['Family'].apply(lambda family: '>3' if family>3 else '<=3' )
test_set['Family'] = test_set['Family'].apply(lambda family: '>3' if family>3 else '<=3')
```

```python
titanic.loc[titanic['Embarked'].isnull(),'Embarked'] = 'S'

test_set['Fare'].fillna(value=test_set['Fare'].median(),inplace=True)
```

```python
# Transformamos los datos

num_attr = ['Age','Fare']
cat_attr = ['Pclass','Sex','Embarked','Deck','Family']
```

```python
full_pipeline = ColumnTransformer([('num',StandardScaler(),num_attr),
                                  ('cat',OneHotEncoder(),cat_attr)])
```

```python
X_train = full_pipeline.fit_transform(titanic)
X_test = full_pipeline.fit_transform(test_set)
y_train = titanic_labels
```

# Primer prueba de modelos
## _SGDClassifier_

```python
sgd = SGDClassifier(random_state=42)
#Primer score
cross_val_score(sgd,X_train,y_train,cv=4,n_jobs=4,
               scoring='accuracy')
```

```python
# Mostramos la matriz de confusión

y_predict = cross_val_predict(sgd,X_train,y_train,cv=4,n_jobs=4)
y_scores = cross_val_predict(sgd,X_train,y_train,cv=4,n_jobs=4,
                            method='decision_function')
conf_mat = confusion_matrix(y_predict, y_train)
plt.imshow(conf_mat)
plt.colorbar();
conf_mat
```

```python
fpr,tpr,thresholds = roc_curve(y_train,y_scores)

plt.figure(figsize=(10,7))
plt.plot(fpr,tpr,linewidth=3)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate');plt.ylabel('True Positive Rate')
plt.grid()
plt.show()
```

```python
roc_auc_score(y_train,y_scores)
```

## _RandomForestClassifier_

```python
rfc = RandomForestClassifier(random_state=42)

y_probas_forest = cross_val_predict(rfc,X_train,y_train,cv=4,n_jobs=4,
                                   method='predict_proba')
```

```python
y_scores_rfc = y_probas_forest[:,1]

fpr_rfc,tpr_rfc,_ = roc_curve(y_train,y_scores_rfc)

plt.figure(figsize=(10,7))
plt.plot(fpr,tpr,linewidth=3,label='SGD')
plt.plot(fpr_rfc,tpr_rfc,linewidth=3,label='Random Forest')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate',size=16);plt.ylabel('True Positive Rate',size=16)
plt.grid();plt.legend(fontsize=16,loc='lower right')
plt.xticks(size=14);plt.yticks(size=14)
plt.show()
```

```python
roc_auc_score(y_train,y_scores_rfc)
```

```python
cross_val_score(rfc,X_train,y_train,cv=5,n_jobs=5,
               scoring='accuracy')
```

## _SupportVectorMachine_

```python
svm = SVC(random_state=42)
y_scores_svm = cross_val_predict(svm,X_train,y_train,cv=4,n_jobs=4,
                            method='decision_function')

fpr_svm,tpr_svm,_ = roc_curve(y_train,y_scores_svm)

plt.figure(figsize=(10,7))
plt.plot(fpr,tpr,linewidth=3,label='SGD')
plt.plot(fpr_rfc,tpr_rfc,linewidth=3,label='Random Forest')
plt.plot(fpr_svm,tpr_svm,linewidth=3,label='Support Vector Machine')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate',size=16);plt.ylabel('True Positive Rate',size=16)
plt.grid();plt.legend(fontsize=16,loc='lower right')
plt.xticks(size=14);plt.yticks(size=14)
plt.show()
```

```python
roc_auc_score(y_train,y_scores_svm)
```

```python
cross_val_score(svm,X_train,y_train,cv=4,n_jobs=4,scoring='accuracy')
```

## _KNeighborsClassifier_

```python
knn = KNeighborsClassifier()

y_probas_knn = cross_val_predict(knn,X_train,y_train,cv=4,n_jobs=4,
                            method='predict_proba')
y_scores_knn = y_probas_knn[:,1]


fpr_knn,tpr_knn,_ = roc_curve(y_train,y_scores_knn)

plt.figure(figsize=(10,7))
plt.plot(fpr,tpr,linewidth=3,label='SGD')
plt.plot(fpr_rfc,tpr_rfc,linewidth=3,label='Random Forest')
plt.plot(fpr_svm,tpr_svm,linewidth=3,label='Support Vector Machine')
plt.plot(fpr_knn,tpr_knn,linewidth=3,label='K Neighbors')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate',size=16);plt.ylabel('True Positive Rate',size=16)
plt.grid();plt.legend(fontsize=16,loc='lower right')
plt.xticks(size=14);plt.yticks(size=14)
plt.show()
```

```python
roc_auc_score(y_train,y_scores_knn)
```

## Resultados en la selección de modelos

# Algo de tuneo de hiperparámetros

```python
#SVM
svm=SVC()
param_grid = [{'C':[1.74,1.75,1.76]},
             {'kernel':['poly','rbf','sigmoid']},
             {'gamma':['scale','auto']}]

gscv_svm = GridSearchCV(svm,param_grid,cv=5,n_jobs=5)

gscv_svm.fit(X_train,y_train)
```

```python
gscv_svm.best_params_
```

```python
# Accuracy final
svm_final = gscv_svm.best_estimator_
cross_val_score(svm_final,X_train,y_train,cv=3,n_jobs=3)
```

# Submission para Kaggle

```python
ids_test = test_set['PassengerId']
survived_test = svm_final.predict(X_test)
```

```python
submission = pd.DataFrame({'PassengerId':ids_test,'Survived':survived_test})
```

```python
submission.to_csv('/mnt/d/ML/kaggle/titanic/submission.csv',index=False)
```

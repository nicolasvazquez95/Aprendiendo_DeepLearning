# Creamos el _test set_

Hay toda una implementación _from scratch_ de cómo se separa la data en 2, de acuerdo a una fracción determinada (pág. 86-88). Yo me voy a quedar con la que ya está hecha en `sklearn`.

```python
from sklearn.model_selection import train_test_split
```

```python
train_set, test_set = train_test_split(housing,test_size=0.2,
                                      random_state=42)
```

Supongamos que sabemos que el ingreso medio es un atributo muy importante para predecir los precios. Quizá sería bueno asegurarse que el test set es representativo de todas las categorías de ingresos en todo el dataset. 

Como es un atributo numérico continuo, necesitamos crear categorías. Mirando mas de cerca el histograma, vemos que casi todos los valores están fuertemente localizados entre 1.5-6. Es importante que ahora que vamos a segmentar los datos, cada segmento tenga una cantidad suficiente de instancias. La implementación del libro usa `pd.cut()`para crear 5 categorías, etiquetadas del uno al 5:

```python
import numpy as np
```

```python
housing["income_cat"] = pd.cut(housing["median_income"],
                                      bins=[0.,1.5,3.0,4.5,6.,np.inf],
                                      labels=[1,2,3,4,5])
```

```python
housing["income_cat"].hist(figsize=(8,6));
```

Con esto hecho, vamos a hacer un sampleo estratificado, basado en la categoría de ingreso. Para eso usamos la clase de `sklearn` `StratifiedShuffleSplit`:

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)

for train_idx,test_idx in split.split(housing,housing["income_cat"]):
    strat_train_set = housing.loc[train_idx]
    strat_test_set = housing.loc[test_idx]
```

Veamos si esto funcionó como esperábamos. El histograma debería quedar proporcional a lo que vimos en el histograma anterior.

```python
strat_test_set["income_cat"].hist(figsize=(8,6));
```

Ahora, lo que deberíamos hacer es devolver los datos a su estado original, removiendo el atributo `income_cat`.

```python
for set_ in (strat_train_set,strat_test_set):
    set_.drop("income_cat",axis=1,inplace=True)
```

Todo esto que hemos hecho nos va a venir bien cuando discutamos la validación cruzada. Es una parte crítica de cualquier proyecto de ML.

## Transformers Custom

Muchas veces va a ser necesario escribir nuestros propios algoritmos de limpieza, o para combinar atributos específicos. Para que funcionen dentro del ecosistema Scikit-Learn, es necesario crear una clase e implementarle tres métodos: `fit()`, `transform()`, y `fit_transform()`.

Podemos usar clases base que tendrán algunas cosas precocinadas, como la `TransformerMixin`, y el `BaseEstimator`.

Como ejemplo, el libro implementa un Transformer que añade los atributos combinados que estuvimos analizando antes:

```python
from sklearn.base import BaseEstimator,TransformerMixin
```

```python
rooms_ix,bedrooms_ix,population_ix,households_ix = 3,4,5,6
```

```python
class CombinedAttributesAdder(BaseEstimator,TransformerMixin):
    def __init__(self,add_bedrooms_per_room=True):
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self,X,y=None):
        return self
    def transform(self,X,y=None):
        rooms_per_household = X[:,rooms_ix]/X[:,households_ix]
        population_per_household = X[:,population_ix]/X[:,households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:,bedrooms_ix]/X[:,rooms_ix]
            return np.c_[X,rooms_per_household,population_per_household,
                        bedrooms_per_room]
        else:
            return np.c_[X,rooms_per_household,population_per_household]
```

```python
attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)
```

Acá hay un hiperparámetro, `add_bedrooms_per_room`, que está seteado `True` por default. 

Siempre se puede añadir un hiperparámetro para testear algún paso en la preparación de los datos del que no estamos 100% seguros.

## Feature Scaling

Con pocas excepciones, los algoritmos no funcionan bien si los atributos numéricos de entrada tienen escalas muy distintos. Por ejemplo, el número de habitaciones va de 6 a 39320, mientras que los ingresos van de 0 a 15. Notemos que escalear los valores objetivo (labels) generalmente no va a ser requerido. 

Hay dos formas de lograr que todos los atributos estén en la misma escala: `min-max scaling` y `standardization`. Ver pág. 109 para una descripción de ambos, sus ventajas y desventajas.

Las clases son `MinMaxScaler` y `StandardScaler`.

## Transformation Pipelines
Para ayudar a ejecutar las transformaciones en el orden correcto, Scikit-Learn tiene la clase `Pipeline` para ayudar con estas secuencias.

El Pipeline de abajo trabaja sobre los atributos numéricos.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
num_pipeline = Pipeline([('imputer',SimpleImputer(strategy='median')),
                        ('attribs_adder',CombinedAttributesAdder()),
                        ('std_scaler',StandardScaler())])

housing_num_tr = num_pipeline.fit_transform(housing_num)
```

Hasta acá, hemos podido manejar las columnas categóricas y las numéricas por separado. Desde la versión 0.2, Scikit-Learn tiene el `ColumnTransformer`, que nos deja manejar todas las columnas a la vez.

Lo vamos a usar para aplicar todas las transformaciones a los datos de `housing`:

```python
from sklearn.compose import ColumnTransformer
```

```python
num_attr = list(housing_num)
cat_attr = ['ocean_proximity']

full_pipeline = ColumnTransformer([('num',num_pipeline,num_attr),
                                  ('cat',OneHotEncoder(),cat_attr),])

housing_prepared = full_pipeline.fit_transform(housing)
```

## Validación cruzada
Vamos a usar `K-fold cross-validation`. Este código divide el training set en diez subsets aleatoriamente, denominados folds. Entonces lo que hace es entrenar y evaluar el modelo 10 veces, eligiendo un fold diferente cada vez para evaluación, y entrenando en los otros 9 folds. El resultado entonces es un array con 10 scores:

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(tree_reg,housing_prepared,housing_labels,
                        scoring='neg_mean_squared_error',cv=10,
                        n_jobs=4,verbose=1)
""" cross_val mide con una función de utilidad y no de costo 
(quiere maximizar), asique ponemos -RMSE para los scores.
"""

tree_rmse_scores = np.sqrt(-scores)

print('Scores: ',tree_rmse_scores)
print('Mean: ',tree_rmse_scores.mean())
print('STD: ',tree_rmse_scores.std())
```

El árbol de decisión no se ve tan bueno como antes. En realidad, incluso  anda peor que el modelo lineal. La validación cruzada nos permite obtener no sólo un estimado de la performance del modelo, sino también sobre qué tan preciso es esta estimación. El árbol de decisión tiene un score $71000 \pm 2000$. La validación cruzada es costosa, puesto que hay que entrenar el modelo varias veces, por lo que no siempre será posible.

Probemos con uno más: `RandomForestRegressor`. Este lo vamos a estudiar en detalle en el capítulo 7, pero básicamente lo que ace es entrenar varios Decision Trees en pequeños subsets de las features, y luego promedia sus predicciones. 

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor()
forest_reg.fit(housing_prepared,housing_labels)

forest_reg_predictions = forest_reg.predict(housing_prepared)
forest_reg_rmse = np.sqrt(mean_squared_error(forest_reg_predictions,
                                             housing_labels))

print('RMSE: ',forest_reg_rmse)
```

```python
#Validación cruzada sobre el modelo RandomForestRegressor
#n_jobs es para paralelizar (joblib)
forest_reg_scores = cross_val_score(forest_reg,housing_prepared,housing_labels,
                                   scoring='neg_mean_squared_error',cv=10,
                                    n_jobs=4,verbose=1)

forest_reg_rmse_scores = np.sqrt(-scores)

print('Scores: ',forest_reg_rmse_scores)
print('Mean: ',forest_reg_rmse_scores.mean())
print('STD: ',forest_reg_rmse_scores.std())
```

Bastante mejor! Random Forests parece prometedor. Sin embargo, notemos que en el score el training set tiene RMSE mucho más bajo que en los test de validación, lo que significa que el modelo todavía esta overfiteando los datos. Posibles soluciones? Simplificar el modelo, regularizarlo, conseguir más datos. Antes que nada deberíamos intentar otros algoritmos ML (como una red neuronal), para no perder mucho tiempo tocando los hiperparámetros. # Tuneo fino del modelo
Búsqueda de los mejores hiperparámetros.
## Búsqueda en grilla 
Con `GridSearchCV`, todo lo que necesitamos es decirle con qué parámetros queremos experimentar y qué valores probar, y usar validación cruzada para evaluar todas las posibles combinaciones de hiperparámetros. 

El siguiente código busca la mejor combinación de hiperparámetros para el `RandomForestRegressor`:

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = [
    {'n_estimators':[20,30,35],'max_features':[4,6,8,10]},
    {'bootstrap':[False],'n_estimators':[10,12],
     'max_features':[2,3,4]},
]
```

```python
forest_reg = RandomForestRegressor()

grid_search = GridSearchCV(forest_reg,param_grid,cv=5,
                          scoring='neg_mean_squared_error',
                          return_train_score=True,
                          n_jobs=4,verbose=2)
```

```python
# Esto demora un montón, asique ojo si se corre
grid_search.fit(housing_prepared,housing_labels)
```

Una vez que hace todas las rondas de entrenamiento, se puede obtener la mejor combinación de hiperparámetros.

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
cvres = grid_search.cv_results_
for mean_score,params in zip(cvres['mean_test_score'],cvres['params']):
    print(np.sqrt(-mean_score),params)
```

## Búsqueda random
La búsqueda por grila es buena cuando estamos explorando relativamente poco parámetros, pero cuando el espacio a cubrir es muy grande, es preferible usar la clase `RandomizedSearchCV`. En vez de intentar por todas las combinaciones posibles, evalúa en un número de combinaciones random para cada hiperparámetro, en cada iteración. Tenemos control sobre el nro. de iteraciones.
## Ensambles
Una alternatia es intentar combinar modelos que funcionan bien, especialmente cuando los modelos individuales tienen errores muy distintos.

## Evaluación del modelo en el Test Set

```python
final_model = grid_search.best_estimator_

X_test = strat_test_set.drop('median_house_value',axis=1)
y_test = strat_test_set['median_house_value'].copy()

X_test_prepared = full_pipeline.transform(X_test)
final_predictions = final_model.predict(X_test_prepared)

final_rmse = np.sqrt(mean_squared_error(y_test,final_predictions))

print('Final RMSE: ', final_rmse)
```

Calculemos el intervalo de confianza a un 95% de este error de generalización:

```python
from scipy import stats
confidence = 0.95

squared_errors = (final_predictions - y_test)**2

np.sqrt(stats.t.interval(confidence,len(squared_errors)-1,
                        loc=squared_errors.mean(),
                        scale=stats.sem(squared_errors)))
```

3. Try adding a transformer in the preparation pipeline to select only the most important attributes

Sacado de la solución del libro.

```python
def indices_of_top_k(arr, k):
    return np.sort(np.argpartition(np.array(arr), -k)[-k:])
```

```python
class TopFeatureSelector(BaseEstimator, TransformerMixin):
    def __init__(self, feature_importances, k):
        self.feature_importances = feature_importances
        self.k = k
    def fit(self, X, y=None):
        self.feature_indices_ = indices_of_top_k(self.feature_importances, self.k)
        return self
    def transform(self, X):
        return X[:, self.feature_indices_]
```

### _Cabin_

[Medium](https://miro.medium.com/max/875/0*krPK8PbfZ_SnvYTg.png) tiene una fotograf√≠a donde se muestra c√≥mo estaban repartidas las cabinas en el barco:


- Se supone que la primera clase estaba localizada en A,B,C;
- Segunda clase ocupaba D y E;
- Tercera clase estaba en F y G.

```python
#train['Cabin'].unique()
```

```python
# Con los elementos de Cabin, armamos una nueva variable Deck, que va a ser m√°s √∫til
## Clase U guarda los NANs
train['Deck'] = train['Cabin'].apply(lambda cabin: cabin[0] if pd.notnull(cabin) else 'U')
```

```python
train.groupby('Deck')['Survived'].mean().plot(kind='bar') # Tasa de supervivencia por deck
```

La [Enciclopedia Titanica](https://www.encyclopedia-titanica.org/cabins.html) dice que 'T' es el √∫nico pasajero que viaj√≥ en la Boat Deck, y muri√≥. Asumiendo que es un pasajero de alta clase, podr√≠amos elegir moverlo a 'U' o a 'A'... Lo pongo en U. Despu√©s vamos a ver qu√© hacemos con todos esos 'U' (o los dejamos as√≠?)

```python
# Qu√© pasa si los agrupamos de acuerdo a la relaci√≥n entre clase y deck ?
train_2 = train.copy()
```

```python
train_2['Deck'] = train_2['Deck'].replace(['A','B','C'],'ABC')
train_2['Deck'] = train_2['Deck'].replace(['D','E'],'DE')
train_2['Deck'] = train_2['Deck'].replace(['F','G'],'FG')
train_2['Deck'] = train_2['Deck'].replace(['U','T'],'U') # Recordar esto!
```

```python
train_2.groupby('Deck')['Survived'].mean().plot(kind='bar') # Tasa de supervivencia por deck
```

```python
#Histogramas de clase de Pasajeros por Deck
#train_2.groupby('Deck')['Pclass'].hist(by=train_2['Deck'])
```

__Decisi√≥n:__ El histograma de U parece contener informaci√≥n que no es aleatoria, sino que los pasajeros de clase 3 son los que m√°s desconocidos muestran. Creo que para no introducir un sesgo, no podemos poner estos datos en ning√∫n lado, y mi decisi√≥n es dejarlos en una categor√≠a aparte. Ac√° no estoy muy seguro, sin embargo, tampoco tengo una idea superadora.

### _Parch/Sibsp_

```python
train.groupby('Parch')['Survived'].mean().plot(kind='bar') #Survival rate Parch
```

```python
train.groupby('SibSp')['Survived'].mean().plot(kind='bar') #Survival rate Parch
```

```python
train_2['SP'] = train['Parch']+train['SibSp']
train_2.groupby('SP')['Survived'].mean().plot(kind='bar')
```

Tenemos 2 opciones ac√°:
- Quedarnos con las 2 variables, y subdividir en dos (4) categor√≠as, para achicar los datos: ponemos corte para Parch y Sib en 3 (es decir, una categor√≠a es de 0 a 3, y la otra de 3 a inf). Es decir, nos quedamos con dos variables, cortadas en 2 c/u.
- Dado lo que se ve en el histograma de Parch+Sibsp, se ve que hay claramente una l√≠nea entre SibSp+Parch>=4 y menores. Si hacemos el corte ah√≠ y descartamos la otra informaci√≥n, nos queda una variable de dos categor√≠as.
